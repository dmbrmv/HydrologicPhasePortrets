```python
import matplotlib.pyplot as plt
import numpy as np


def cool_plot():
    """Generate a sine wave plot using the global variable 'b' as the frequency multiplier.

    The function creates a plot of y = sin(x * b) where 'b' is a predefined variable.
    """
    x = np.linspace(0, 10, 100)
    y = np.sin(x * b)  # Using the variable 'b'
    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label=f"Sine wave with frequency {b}")
    plt.title("Cool Plot")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.legend()
    plt.grid(True)
    plt.show()


cool_plot()
```
